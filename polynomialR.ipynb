```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Dataset/Position_Salaries.csv')
dataset
```

```python
dataset.shape
```

```python
X = dataset.iloc[:,1:2].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol_reg = PolynomialFeatures(degree=4)  # change degree for best fitting
x_poly = pol_reg.fit_transform(X)
x_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,Y)
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title('Truth or Bluff (linear regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg2.predict(pol_reg.fit_transform(X)),color='blue')
plt.title('Truth or Bluff (Polynomial regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

POLYNOMIAL ON RANDOM DATA

```python
x_grid = np.arange(min(X),max(X),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
x_grid
```

```python
plt.scatter(X,Y,color='red')
plt.plot(x_grid,lin_reg2.predict(pol_reg.fit_transform(x_grid)),color='blue')
plt.title('Truth or Bluff (Polynomial regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
